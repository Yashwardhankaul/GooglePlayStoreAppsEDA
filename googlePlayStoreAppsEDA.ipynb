I will make a function for horizontal box plots so that I can use it in the future.

```
def horizontal_box_plot(h,b,t,x,y):
  """
  h = Height of the unique labels
  b = Different labels
  t = Title
  x = Xlabel
  y = Ylabel
  """
  figure(num=0, figsize=(8, 6), dpi=100, facecolor='w', edgecolor='k')
  height = h
  bars = b
  y_pos = np.arange(len(bars))

  # Add title and axis names
  plt.title(t)
  plt.xlabel(x)
  plt.ylabel(y)

  # Create horizontal bars
  plt.barh(y_pos, height)

  # Create names on the y-axis
  plt.yticks(y_pos, bars)
 
  # Show graphic
  plt.show()
```

Functions for getting lists of frequencies of labels and the list of Category names.

```
def get_freq(dataFrame):
  freq = dataFrame.value_counts().tolist()
  return freq

def get_cat(dataFrame):
  cat = dataFrame.value_counts().keys().tolist()
  return cat
```

```
horizontal_box_plot(get_freq(df1['Category']),get_cat(df1['Category']),'Bar Plot for unique App Categories','Number of Apps','Categories')
```

Will have to clean this data up for visualization

```
df1['Installs'] = df1['Installs'].str.replace('+', ' ')
df1['Installs'] = df1['Installs'].str.replace(',','')
df1['Installs'] = df1['Installs'].replace('Free','0')
pd.to_numeric(df1['Installs'])
```

```
def get_label_num_sum(col, label, toAdd): 
  """
  col = Name of column you want to filter through
  label = Label you want to filter
  toAdd = column name of numerical value you want to add
  
  Returns - int64
  """
  t = df1[df1[col] == label ]
  p = t[toAdd]
  p = pd.to_numeric(p)
  p1=p.sum() 
  return p1
```

```
categories = get_cat(df1['Category'])
a = []
for items in categories:
  s = get_label_num_sum('Category',items,'Installs')
  a.append(s)
```

To better understand the difference in numbers of 'Free' and 'Paid' Apps in the different app categories I will create a nested pie chart.

```
def grp_size(item):
  h = df1[df1['Category'] == item]
  v = h.Type.value_counts().tolist()
  return sum(v)
```

```
def sub_grp_size(item,i):
  h = df1[df1['Category'] == item]
  v = h.Type.value_counts().tolist()
  return v[i]
```

```
group_names=['GAME','COMMUNICATION','PRODUCTIVITY','SOCIAL']
group_size=[grp_size('GAME'),grp_size('COMMUNICATION'),grp_size('PRODUCTIVITY'),grp_size('SOCIAL')]
subgroup_names=['FREE', 'PAID', 'FREE', 'PAID', 'FREE', 'PAID', 'FREE', 'PAID']
subgroup_size=[sub_grp_size('GAME',0),sub_grp_size('GAME',1),sub_grp_size('COMMUNICATION',0),sub_grp_size('COMMUNICATION',1),sub_grp_size('PRODUCTIVITY',0),sub_grp_size('PRODUCTIVITY',1),sub_grp_size('SOCIAL',0),sub_grp_size('SOCIAL',1)]
a, b, c, d =[plt.cm.Blues, plt.cm.Reds, plt.cm.Greens, plt.cm.Purples]
 
# First Ring (outside)
fig, ax = plt.subplots()
ax.axis('equal')
mypie, _ = ax.pie(group_size, radius=2.4, labels=group_names, colors=[a(0.6),b(0.6),c(0.6),d(0.6)] )
plt.setp( mypie, width=0.3, edgecolor='white')

# Second Ring (Inside)
mypie2, _ = ax.pie(subgroup_size, radius=2.4-0.3, labels=subgroup_names, labeldistance=0.7, colors=[a(0.5), a(0.4), b(0.5), b(0.4), c(0.5), c(0.4),d(0.5), d(0.4)])
plt.setp( mypie2, width=0.4, edgecolor='white')
plt.margins(0,0)
 
# show it
plt.show()
```
